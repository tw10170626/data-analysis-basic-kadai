```python
import matplotlib.pyplot as plt
import numpy as np

fig, axes = plt.subplots(2, 3,tight_layout=True )

y0 = [1, 2, -5, 2]
axes[0,0].plot(y0)

x0 = np.random.normal(15, 5, 2000)
axes[0,2].hist(x0)

z = np.linspace(0, 10 ,30)
z0 = 2 + 2 * np.sin(2 * z)

axes[0,1].plot(z, z0, linewidth=2.0)

np.random.seed(3)
a0 = 4 + np.random.normal(0, 2, 60)
b0 = 4 + np.random.normal(0,2,len(a0))

axes[1,0].scatter(a0, b0)



plt.show()
```
